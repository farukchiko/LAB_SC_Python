```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
#x = [0,1,2]
#y = [1,3,2]

x = [1, 3, 5, 7]
y = [2, 4, 6, 8]

f = interp1d(x,y)
y_hat = f(6)
print(y_hat)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x, y, '-ob')
plt.plot(6, y_hat, 'or')
plt.title('Interpolated function at x = 6')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
